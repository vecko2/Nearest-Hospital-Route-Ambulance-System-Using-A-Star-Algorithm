#**SHORTEST PATH**

1. Model ini menggunakan pustaka osmnx untuk mengubah peta kota (dalam hal ini Surabaya) menjadi struktur matematika yang disebut Graf.
*   Node (Titik): Merepresentasikan persimpangan jalan atau ujung jalan.
*   Edge (Sisi): Merepresentasikan ruas jalan yang menghubungkan antar persimpangan, lengkap dengan atribut jarak (meter).  


2. Penentuan Titik Terdekat (Nearest Node)
Karena koordinat pasien dan rumah sakit seringkali tidak tepat berada di tengah persimpangan, model akan mencari node terdekat pada graf dari koordinat asli. Ini memastikan algoritma mulai berjalan dari jaringan jalan yang valid.  


3. Logika Algoritma A* (A-Star)
*   g(n) (Cost so far): Jarak sebenarnya yang telah ditempuh dari lokasi pasien ke node saat ini.
*   h(n) (Heuristic): Estimasi jarak dari node saat ini ke rumah sakit tujuan (biasanya menggunakan jarak garis lurus atau Euclidean distance).
*   f(n): Total bobot node. Model akan selalu memilih node dengan nilai f(n) terkecil.  


4. Proses Eksekusi
*   Model memeriksa tetangga dari node saat ini.
*   Menghitung biaya untuk tiap tetangga menggunakan rumus di atas.
*   Memilih tetangga "paling menjanjikan" (yang paling mendekati arah rumah sakit).
*   Proses ini berulang hingga node rumah sakit tercapai. Inilah mengapa A* lebih cepat daripada algoritma Dijkstra biasa, karena A* "tahu" arah mana yang harus diprioritaskan.


5. Rekonstruksi Rute dan Kalkulasi Jarak
Setelah node tujuan ditemukan, model akan menarik garis balik melalui node-node yang terpilih tadi.
*   Nama Rumah Sakit: Diambil dari metadata rs_data.
*   Jarak Total: Merupakan akumulasi dari panjang edges (ruas jalan) yang dilewati, bukan sekadar jarak tarik garis lurus, sehingga hasilnya sangat akurat dengan kondisi lapangan.






```python
import osmnx as ox
import networkx as nx
import folium
import math

# Filter RS sesuai tipe
rs_filtered = {nama: v for nama, v in rs_data.items() if v[1] == tipe}

# Buat graf dan node pasien
G = ox.graph_from_point(Pasien, dist=10000, network_type='drive')
orig = ox.distance.nearest_nodes(G, Pasien[1], Pasien[0])

# Cari RS terdekat (Euclidean)
min_dist = float('inf')
for nama, (loc, _) in rs_filtered.items():
    node = ox.distance.nearest_nodes(G, loc[1], loc[0])
    dist = math.hypot(loc[1] - Pasien[1], loc[0] - Pasien[0])
    if dist < min_dist:
        min_dist = dist
        Rumah_sakit = loc
        dest = node
        nama_rs = nama

# Heuristik A*
def heuristic(u, v):
    y1, x1 = G.nodes[u]['y'], G.nodes[u]['x']
    y2, x2 = G.nodes[v]['y'], G.nodes[v]['x']
    return math.hypot(x2 - x1, y2 - y1)

# Rute A*
route = nx.astar_path(G, orig, dest, heuristic=heuristic, weight='length')
coords = [(G.nodes[n]['y'], G.nodes[n]['x']) for n in route]

# Peta
m = folium.Map(location=Pasien, zoom_start=13)
folium.PolyLine(coords, color='blue', weight=5, tooltip="A* Path").add_to(m)
folium.Marker(Pasien, popup='Pasien', icon=folium.Icon(color='red')).add_to(m)
folium.Marker(Rumah_sakit, popup=nama_rs, icon=folium.Icon(color='green')).add_to(m)

# Hitung total panjang rute
length = sum(G[u][v][k]['length'] for u, v, k in zip(route[:-1], route[1:], [0]*len(route)))
print(f"Rute node:", route)
print(f"Nama rumah sakit: {nama_rs}")
print(f"Jarak total: {length/1000:.2f} km")
m
```

#**NODE**

Visualisasi node pada peta ketika model mencari jalan tercepat.

```python
import heapq

visited_nodes = set()

def custom_astar_path(G, source, target, heuristic=None, weight='weight'):
    visited_nodes.clear()
    queue = []
    heapq.heappush(queue, (0, source))
    parent = {source: None}
    distance = {source: 0}

    while queue:
        _, current = heapq.heappop(queue)
        if current in visited_nodes:
            continue
        visited_nodes.add(current)

        if current == target:
            break

        for neighbor in G.neighbors(current):
            new_cost = distance[current] + G[current][neighbor][0][weight]
            if neighbor not in distance or new_cost < distance[neighbor]:
                distance[neighbor] = new_cost
                priority = new_cost + heuristic(neighbor, target)
                heapq.heappush(queue, (priority, neighbor))
                parent[neighbor] = current

    # Rekonstruksi rute
    current = target
    path = []
    while current is not None:
        path.append(current)
        current = parent[current]
    path.reverse()
    return path

# Gunakan custom_astar_path
route = custom_astar_path(G, orig, dest, heuristic=heuristic, weight='length')
coords = [(G.nodes[n]['y'], G.nodes[n]['x']) for n in route]

# Tampilkan di peta
m = folium.Map(location=Pasien, zoom_start=12)
folium.PolyLine(coords, color='red', weight=10, tooltip="A* Route").add_to(m)

# Tambahkan semua node yang dikunjungi
for node in visited_nodes:
    lat, lon = G.nodes[node]['y'], G.nodes[node]['x']
    folium.CircleMarker(location=(lat, lon), radius=2, color='blue', fill=True).add_to(m)

folium.Marker(Pasien, popup='Pasien', icon=folium.Icon(color='red')).add_to(m)
folium.Marker(Rumah_sakit, popup=nama_rs, icon=folium.Icon(color='green')).add_to(m)

print(f'total node dikunjungi: {len(visited_nodes)}, {visited_nodes}')
print(f'total node menuju RS: {len(route)}, {route}')
m
```
